```python
!pip install pulp
import pandas as pd
import pulp
import matplotlib.pyplot as plt
import math
import os
import numpy as np

# === CONFIGURACIÓN GENERAL ===
CSV_PATH = "Solar_Energy_Generation.csv"   # Ruta del archivo Kaggle
OUTPUT_DIR = "volcanes_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Parámetros del sistema
N_PANELS = 4
INITIAL_SOC_FRAC = 1  # batería inicia al 50%
M = 1e6                  # Big-M
DAYS = [0]               # un solo día
B_count = 4              # número de baterías


# Capacidad baterías (Wh)
Cmin = 288.0
Cmax = 576.0

# Dispositivos del sistema (del paper)
TYPES = [
    ('EX', 20, 4, 1),   # exterior bulbs
    ('BSC',20, 4, 1),   # classroom bulbs
    ('PC', 60, 5, 2),   # computers
    ('TV',100,1, 2),
    ('TA',50, 1, 2),
    ('BH',20, 1, 1)
]
```

Carga y preparación de datos

```python

# === CONFIGURACIÓN ===
CSV_PATH = "Solar_Energy_Generation.csv"

# === CARGA Y LIMPIEZA ===
df = pd.read_csv(CSV_PATH)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df = df.dropna(subset=["SolarGeneration"]).sort_values("Timestamp").reset_index(drop=True)

# === IDENTIFICAR TODOS LOS DÍAS DISPONIBLES ===
df['Date'] = df['Timestamp'].dt.date
dias = df['Date'].unique()
print("Días encontrados:", len(dias))

# === BUSCAR EL PRIMER DÍA COMPLETO CON 24 HORAS ===
df_day = None
for d in dias:
    temp = df[df['Date'] == d]
    hourly = temp.resample("h", on="Timestamp")["SolarGeneration"].sum()
    if len(hourly) >= 24:
        df_day = hourly.head(24).reset_index()
        break

# Si ningún día tiene 24 horas completas, tomar el primero y rellenar
if df_day is None:
    print("⚠️ Ningún día tiene 24 horas completas, se usará el primero encontrado con interpolación.")
    temp = df[df['Date'] == dias[0]]
    df_hour = temp.resample("h", on="Timestamp")["SolarGeneration"].sum()
    df_hour = df_hour.reindex(pd.date_range(df_hour.index.min().floor('D'),
                                            periods=24, freq='H'), fill_value=None)
    df_day = df_hour.interpolate().reset_index()

# === RENOMBRAR COLUMNAS ===
df_day.columns = ["Timestamp", "SolarGeneration"]

# === ASEGURAR 24 HORAS ===
df_day = df_day.head(24)
df_day["SolarGeneration"] = df_day["SolarGeneration"].fillna(0)

# === CREAR VECTORES OI ===
OI_1panel = df_day["SolarGeneration"].tolist()
OI = [x * N_PANELS for x in OI_1panel]
J = list(range(24))

print(f"Perfil solar listo ({len(OI_1panel)} valores horarios). Día usado: {df_day['Timestamp'].iloc[0].date()}")

# === GRAFICAR PERFIL SOLAR ===
plt.figure(figsize=(10,4))
plt.plot(df_day["Timestamp"], OI_1panel, marker='o', label="1 panel")
plt.plot(df_day["Timestamp"], OI, marker='o', linestyle='--', label="4 paneles")
plt.title(f"Energía solar (día {df_day['Timestamp'].iloc[0].date()}) — 24 horas garantizadas")
plt.ylabel("Energía (Wh)")
plt.xlabel("Hora")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
K = []
for code, demand, qty, pri in TYPES:
    for q in range(1, qty+1):
        K.append({'id': f"{code}{q}", 'demand': demand, 'priority': pri})

Nk = [k['demand'] for k in K]
Pk = [k['priority'] for k in K]
K_count = len(K)

print(f"Total de dispositivos: {K_count}")
pd.DataFrame(K)
```

Creación del modelo

```python
import pulp

model = pulp.LpProblem("Volcanes_Deterministic", pulp.LpMaximize)

# === Variables ===
# Se mantienen tus variables originales
PB = pulp.LpVariable.dicts("PB", (DAYS, J, list(range(B_count))), lowBound=0)
BD = pulp.LpVariable.dicts("BD", (DAYS, J, list(range(B_count))), lowBound=0)
C  = pulp.LpVariable.dicts("C",  (DAYS, J, list(range(B_count))), lowBound=0, upBound=Cmax)
Q  = pulp.LpVariable.dicts("Q",  (DAYS, J, list(range(B_count))), lowBound=0, upBound=1, cat='Binary')
S  = pulp.LpVariable.dicts("S",  (DAYS, J), lowBound=0)
X  = pulp.LpVariable.dicts("X",  (list(range(K_count)), DAYS, J), lowBound=0, upBound=1, cat='Binary')

# --- NUEVA VARIABLE (CRÍTICA) ---
# PD = Energía del Panel que va directo a las Cargas (Dispositivos)
PD = pulp.LpVariable.dicts("PD", (DAYS, J), lowBound=0)

# D_max se elimina, ya que causaba la falla lógica.

# === Restricciones ===
for i in DAYS:
    for j in J:
        # Suministro solar de la hora j
        Oij = OI[j]

        # Eij se elimina, su lógica era incorrecta.

        # Restricciones de Batería (Ecuaciones 1, 2, 3, 5 del paper)
        # Esto se mantiene igual que tu código original.
        for b in range(B_count):
            # (1) Carga mínima para que la batería se considere "activa" (Q=1)
            model += C[i][j][b] >= Cmin * Q[i][j][b], f"C_min_activa_{i}_{j}_{b}"

            # (2) Límite de carga (del paper)
            model += C[i][j][b] <= Cmin * (Q[i][j][b] + 1), f"C_limite_paper_{i}_{j}_{b}"

            # (3) Solo se puede descargar (BD) si la batería está "activa" (Q=1)
            model += BD[i][j][b] <= Q[i][j][b] * M, f"BD_limite_Q_{i}_{j}_{b}"

            # (5) Ecuación de balance de carga de la batería (SoC)
            if j == 0:
                Cprev = INITIAL_SOC_FRAC * Cmax
                model += C[i][j][b] == (Cprev - BD[i][j][b] + PB[i][j][b]), f"SoC_j0_{i}_{j}_{b}"
            else:
                model += C[i][j][b] == (C[i][j-1][b] - BD[i][j][b] + PB[i][j][b]), f"SoC_transicion_{i}_{j}_{b}"


        # --- RESTRICCIONES CORREGIDAS (Arreglan Falla 1 y 2) ---

        # 1. (NUEVA) Balance de Energía Solar:
        # La energía solar (Oij) se divide entre cargar baterías (PB) O ir a cargas (PD).
        # No se puede usar más energía de la que genera el panel.
        sumPB = pulp.lpSum([PB[i][j][b] for b in range(B_count)])
        model += (sumPB + PD[i][j]) <= Oij, f"Balance_Solar_{i}_{j}"

        # 2. (CORREGIDA) Suministro Total S[i][j] - (Reemplaza la lógica de Eij y Eq. 6)
        # El suministro total (S) es lo que sale de las baterías (BD) + lo que va directo del panel (PD).
        sumBD = pulp.lpSum([BD[i][j][b] for b in range(B_count)])
        model += S[i][j] == (sumBD + PD[i][j]), f"Suministro_Total_{i}_{j}"

        # Ecuación (8) del paper - Límite de Suministro (se mantiene)
        # (Usando Nk[k] estático, como en tu setup de datos)
        model += pulp.lpSum([X[k][i][j] * Nk[k] for k in range(K_count)]) <= S[i][j], f"Limite_Suministro_{i}_{j}"

# === Objetivo ===
# (Ecuación 10) - Se mantiene igual, usando tus listas estáticas Nk y Pk
model += pulp.lpSum([X[k][i][j] * Nk[k] * Pk[k]
                     for k in range(K_count) for i in DAYS for j in J]), "Max_served"

print("Modelo construido correctamente.")
```

Solución del modelo

```python
solver= pulp.PULP_CBC_CMD(timeLimit=10)
model.solve(solver)

print("\n--- Resultados de la Optimización ---")
print(f"Estado del Modelo: {pulp.LpStatus[model.status]}")
    # Nota: Si el estado es 'Feasible' o 'Optimal', encontró una solución.
if pulp.LpStatus[model.status] in ["Optimal", "Feasible"]:
     print(f"Valor objetivo final (servido ponderado): {pulp.value(model.objective):.2f}")
else:
     print("El solver terminó sin encontrar una solución válida en el tiempo límite.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Extraer datos detallados de la variable 'X' ---
# (Asegúrate de que el modelo se haya resuelto y 'K_count' y 'J' existan)

# Esta lista almacenará cada registro individual (dispositivo, hora, estado)
datos_cronograma = []

# Iteramos por cada dispositivo (k) y cada hora (j)
for k in range(K_count):
    # Obtenemos el ID del dispositivo (ej: "PC1") desde la lista K
    device_id = K[k]['id']

    for j in J:
        # Obtenemos el valor de la variable X[k][0][j] (0 o 1)
        # Usamos int(pulp.value(...) > 0.5) para convertir 1.0 a 1
        estado = int(pulp.value(X[k][0][j]) > 0.5)

        datos_cronograma.append({
            "dispositivo": device_id,
            "hora": j,
            "estado": estado
        })

# Convertimos la lista a un DataFrame "largo"
df_largo = pd.DataFrame(datos_cronograma)

# --- 2. Crear la Tabla Pivote (Matriz Dispositivo vs. Hora) ---

# Pivotamos el DataFrame:
# - índice = dispositivos
# - columnas = horas
# - valores = estado (0 o 1)
df_cronograma_matriz = df_largo.pivot(
    index="dispositivo",
    columns="hora",
    values="estado"
)

# (Opcional) Reordenamos las filas para que coincidan con el orden de la imagen
# La lista 'K' ya tiene el orden lógico
orden_dispositivos = [d['id'] for d in K]
df_cronograma_matriz = df_cronograma_matriz.reindex(orden_dispositivos)

# --- 3. Mostrar la Tabla ---
print("--- Cronograma de Activación (Tabla) ---")
print("1 = ENCENDIDO, 0 = APAGADO")
display(df_cronograma_matriz)


# --- 4. Generar el Mapa de Calor (Heatmap) ---
print("\n--- Cronograma de Activación (Gráfico) ---")

datos_heatmap = df_cronograma_matriz.values
etiquetas_y_dispositivos = df_cronograma_matriz.index
etiquetas_x_horas = df_cronograma_matriz.columns

fig, ax = plt.subplots(figsize=(16, 8)) # Ajusta el tamaño si es necesario

# Usamos pcolormesh para crear el heatmap con bordes
# Necesitamos N+1 bordes para N celdas
x_edges = np.arange(len(etiquetas_x_horas) + 1)
y_edges = np.arange(len(etiquetas_y_dispositivos) + 1)

im = ax.pcolormesh(x_edges, y_edges, datos_heatmap, cmap='viridis', edgecolors='white', linewidth=1)

# Configuramos las etiquetas en el centro de las celdas
ax.set_xticks(np.arange(len(etiquetas_x_horas)) + 0.5)
ax.set_yticks(np.arange(len(etiquetas_y_dispositivos)) + 0.5)

ax.set_xticklabels(etiquetas_x_horas)
ax.set_yticklabels(etiquetas_y_dispositivos)

# Invertimos el eje Y para que los dispositivos aparezcan en el orden de la tabla
ax.invert_yaxis()

# Añadimos la barra de color
cbar = fig.colorbar(im, ax=ax, ticks=[0, 1])
cbar.ax.set_yticklabels(['APAGADO (0)', 'ENCENDIDO (1)'])

# Títulos y etiquetas
ax.set_title("Cronograma de Activación de Dispositivos por Hora", fontsize=16)
ax.set_xlabel("Hora del Día (0-23)", fontsize=12)
ax.set_ylabel("Dispositivo", fontsize=12)

plt.tight_layout()
plt.show()
```
