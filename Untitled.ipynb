```python
import pandas as pd
import pulp
import matplotlib.pyplot as plt
import math
import os
import numpy as np

# === CONFIGURACIÓN GENERAL ===
CSV_PATH = "Solar_Energy_Generation.csv"   # Ruta del archivo Kaggle
OUTPUT_DIR = "volcanes_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Parámetros del sistema
N_PANELS = 4
INITIAL_SOC_FRAC = 0.5   # batería inicia al 50%
M = 1e6                  # Big-M
DAYS = [0]               # un solo día
B_count = 4              # número de baterías

# Capacidad baterías (Wh)
Cmin = 288.0
Cmax = 576.0

# Dispositivos del sistema (del paper)
TYPES = [
    ('EX', 20, 4, 1),   # exterior bulbs
    ('BSC',20, 4, 1),   # classroom bulbs
    ('PC', 60, 5, 2),   # computers
    ('TV',100,1, 2),
    ('TA',50, 1, 2),
    ('BH',20, 1, 1)
]
```

Carga y preparación de datos

```python

# === CONFIGURACIÓN ===
CSV_PATH = "Solar_Energy_Generation.csv"
N_PANELS = 4

# === CARGA Y LIMPIEZA ===
df = pd.read_csv(CSV_PATH)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df = df.dropna(subset=["SolarGeneration"]).sort_values("Timestamp").reset_index(drop=True)

# === IDENTIFICAR TODOS LOS DÍAS DISPONIBLES ===
df['Date'] = df['Timestamp'].dt.date
dias = df['Date'].unique()
print("Días encontrados:", len(dias))

# === BUSCAR EL PRIMER DÍA COMPLETO CON 24 HORAS ===
df_day = None
for d in dias:
    temp = df[df['Date'] == d]
    hourly = temp.resample("H", on="Timestamp")["SolarGeneration"].sum()
    if len(hourly) >= 24:
        df_day = hourly.head(24).reset_index()
        break

# Si ningún día tiene 24 horas completas, tomar el primero y rellenar
if df_day is None:
    print("⚠️ Ningún día tiene 24 horas completas, se usará el primero encontrado con interpolación.")
    temp = df[df['Date'] == dias[0]]
    df_hour = temp.resample("H", on="Timestamp")["SolarGeneration"].sum()
    df_hour = df_hour.reindex(pd.date_range(df_hour.index.min().floor('D'),
                                            periods=24, freq='H'), fill_value=None)
    df_day = df_hour.interpolate().reset_index()

# === RENOMBRAR COLUMNAS ===
df_day.columns = ["Timestamp", "SolarGeneration"]

# === ASEGURAR 24 HORAS ===
df_day = df_day.head(24)
df_day["SolarGeneration"] = df_day["SolarGeneration"].fillna(0)

# === CREAR VECTORES OI ===
OI_1panel = df_day["SolarGeneration"].tolist()
OI = [x * N_PANELS for x in OI_1panel]
J = list(range(24))

print(f"Perfil solar listo ({len(OI_1panel)} valores horarios). Día usado: {df_day['Timestamp'].iloc[0].date()}")

# === GRAFICAR PERFIL SOLAR ===
plt.figure(figsize=(10,4))
plt.plot(df_day["Timestamp"], OI_1panel, marker='o', label="1 panel")
plt.plot(df_day["Timestamp"], OI, marker='o', linestyle='--', label="4 paneles")
plt.title(f"Energía solar (día {df_day['Timestamp'].iloc[0].date()}) — 24 horas garantizadas")
plt.ylabel("Energía (Wh)")
plt.xlabel("Hora")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
K = []
for code, demand, qty, pri in TYPES:
    for q in range(1, qty+1):
        K.append({'id': f"{code}{q}", 'demand': demand, 'priority': pri})

Nk = [k['demand'] for k in K]
Pk = [k['priority'] for k in K]
K_count = len(K)

print(f"Total de dispositivos: {K_count}")
pd.DataFrame(K)
```

Creación del modelo

```python
model = pulp.LpProblem("Volcanes_Deterministic", pulp.LpMaximize)

# Variables (acceso anidado)
PB = pulp.LpVariable.dicts("PB", (DAYS, J, list(range(B_count))), lowBound=0)
BD = pulp.LpVariable.dicts("BD", (DAYS, J, list(range(B_count))), lowBound=0)
C  = pulp.LpVariable.dicts("C",  (DAYS, J, list(range(B_count))), lowBound=0, upBound=Cmax)
Q  = pulp.LpVariable.dicts("Q",  (DAYS, J, list(range(B_count))), lowBound=0, upBound=1, cat='Binary')
S  = pulp.LpVariable.dicts("S",  (DAYS, J), lowBound=0)
X  = pulp.LpVariable.dicts("X",  (list(range(K_count)), DAYS, J), lowBound=0, upBound=1, cat='Binary')

D_max = sum(Nk)

# Restricciones
for i in DAYS:
    for j in J:
        Oij = OI[j]
        Eij = 1 if D_max > Oij else 0

        for b in range(B_count):
            model += C[i][j][b] >= Cmin * Q[i][j][b]
            model += C[i][j][b] <= Cmin * (Q[i][j][b] + 1)
            model += BD[i][j][b] <= Q[i][j][b] * M

            if j == 0:
                Cprev = INITIAL_SOC_FRAC * Cmax
                model += C[i][j][b] == (Cprev - BD[i][j][b] + PB[i][j][b])
            else:
                model += C[i][j][b] == (C[i][j-1][b] - BD[i][j][b] + PB[i][j][b])

        sumBD = pulp.lpSum([BD[i][j][b] for b in range(B_count)])
        sumPB = pulp.lpSum([PB[i][j][b] for b in range(B_count)])

        if Eij == 1:
            model += S[i][j] == (sumBD + Oij)
        else:
            model += S[i][j] == (Oij - sumPB)

        model += pulp.lpSum([X[k][i][j] * Nk[k] for k in range(K_count)]) <= S[i][j]

# Objetivo: maximizar energía servida ponderada por prioridad
model += pulp.lpSum([X[k][i][j] * Nk[k] * Pk[k]
                     for k in range(K_count) for i in DAYS for j in J]), "Max_served"
print("Modelo construido correctamente.")
```

Solución del modelo

```python
solver = pulp.PULP_CBC_CMD(msg=1, timeLimit=120, threads=4)
model.solve(solver)

print("Estado:", pulp.LpStatus[model.status])
print("Valor objetivo:", pulp.value(model.objective))
```

```python
# Energía disponible S[i][j]
E_served = [pulp.value(S[0][j]) for j in J]

# Estado de carga
soc_data = []
for b in range(B_count):
    for j in J:
        soc_data.append({"bateria": b, "hora": j, "SoC_Wh": pulp.value(C[0][j][b])})
soc_df = pd.DataFrame(soc_data)

# Cargas encendidas
schedule = []
for j in J:
    activos = [K[k]['id'] for k in range(K_count) if pulp.value(X[k][0][j]) > 0.5]
    schedule.append({"hora": j, "ON": ", ".join(activos)})
schedule_df = pd.DataFrame(schedule)

# Mostrar primeros resultados
display(soc_df.head(), schedule_df.head())
```
